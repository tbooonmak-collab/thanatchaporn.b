## 3. Shape และมิติของ Array

**Shape** คือ "รูปร่าง" ของ array บอกว่ามีกี่แถว กี่คอลัมน์ (หรือกี่มิติ) — เป็นสิ่งที่ต้องเช็คบ่อยที่สุดเวลาทำงานกับ NumPy เพราะการคำนวณหลายอย่างต้องการ shape ที่ตรงกัน

| Attribute | ความหมาย |
|---|---|
| `.shape` | รูปร่างของ array เป็น tuple (เชื่อมกับ tuple ที่เรียนมา!) |
| `.ndim` | จำนวนมิติ (1D, 2D, 3D, ...) |
| `.size` | จำนวนสมาชิกทั้งหมด |
| `.dtype` | ชนิดข้อมูลของสมาชิกใน array |


````python
scores_1d = np.array([78, 85, 62, 90, 55])

print("shape:", scores_1d.shape)    # (5,) -- tuple ที่มีค่าเดียว (1 มิติ มี 5 ตัว)
print("ndim:", scores_1d.ndim)      # 1
print("size:", scores_1d.size)      # 5
print("dtype:", scores_1d.dtype)    # int64 (หรือ int32 ขึ้นกับระบบ)
````

````python
scores_2d = np.array([
    [80, 90, 75],
    [60, 70, 65],
    [95, 88, 92],
])

print("shape:", scores_2d.shape)    # (3, 3) -- 3 แถว 3 คอลัมน์
print("ndim:", scores_2d.ndim)      # 2
print("size:", scores_2d.size)      # 9 (3x3)

# unpack shape ด้วย tuple unpacking (เชื่อมกับเรื่อง tuple ที่เรียนมา)
rows, cols = scores_2d.shape
print(f"มี {rows} แถว, {cols} คอลัมน์")
````

### 3.1 `reshape()` — เปลี่ยนรูปร่าง array โดยไม่เปลี่ยนข้อมูล

ใช้บ่อยมากเวลาต้องปรับรูปร่างข้อมูลให้เข้ากับฟังก์ชัน/โมเดลที่ต้องการ shape เฉพาะแบบ


````python
numbers = np.arange(1, 13)   # array ของเลข 1-12 (1 มิติ มี 12 ตัว)
print("ก่อน reshape:", numbers)
print("shape ก่อน:", numbers.shape)

````

````python
# เปลี่ยนเป็น 3 แถว 4 คอลัมน์ (3 x 4 = 12 พอดี ต้องคูณได้เท่ากับจำนวนสมาชิกเดิม)
numbers_2d = numbers.reshape(3, 4)
print("หลัง reshape เป็น (3,4):")
print(numbers_2d)
````

````python
# เปลี่ยนเป็น 4 แถว 3 คอลัมน์ก็ได้ (4 x 3 = 12 เท่ากัน)
numbers_2d_v2 = numbers.reshape(4, 3)
print("หลัง reshape เป็น (4,3):")
print(numbers_2d_v2)


# ลอง reshape เป็นขนาดที่คูณกันไม่ได้ 12 -> จะ error
# numbers.reshape(3, 5)   # ลบ # แล้วรันดู จะเห็น ValueError
````

## 4. Vectorized Operation (การคำนวณแบบเวกเตอร์)

นี่คือจุดเด่นที่สุดของ NumPy — คำนวณกับสมาชิก**ทุกตัวใน array พร้อมกัน** ในคำสั่งเดียว โดยไม่ต้องเขียน loop เลย เรียกว่า **vectorized operation**

### 4.1 คำนวณกับค่าคงที่ (Scalar)


````python
scores = np.array([78, 85, 62, 90, 55])

# เทียบกับ list ที่บวกตรงๆไม่ได้ (ต้องใช้ comprehension) -- array ทำได้เลย!
print("บวก 5 ทุกตัว:", scores + 5)
print("คูณ 2 ทุกตัว:", scores * 2)
print("หารด้วย 10 ทุกตัว:", scores / 10)
print("ยกกำลัง 2 ทุกตัว:", scores ** 2)

# เทียบกับวิธีเดิมด้วย list comprehension
scores_list = [78, 85, 62, 90, 55]
scores_bonus_list = [s + 5 for s in scores_list]   # ต้องเขียน comprehension
scores_bonus_array = scores + 5                      # array ทำได้บรรทัดเดียว ไม่ต้อง loop
print(scores_bonus_list)
print(scores_bonus_array)
````

### 4.2 คำนวณระหว่าง array กับ array (element-wise)

เมื่อ array 2 ตัวมี shape ตรงกัน การคำนวณจะทำ "ทีละตำแหน่งที่ตรงกัน" (element-wise) อัตโนมัติ


````python
midterm = np.array([78, 85, 62, 90, 55])
final = np.array([82, 80, 70, 95, 60])

# บวกกันแบบ element-wise (ตำแหน่งที่ 0 บวกกับตำแหน่งที่ 0, ตำแหน่ง 1 บวกกับตำแหน่ง 1, ...)
total = midterm + final
print("รวมคะแนน:", total)

# หาค่าเฉลี่ยของแต่ละคน (เทียบกับ Notebook ก่อนหน้าที่ใช้ zip() กับ list ธรรมดา)
average = (midterm + final) / 2
print("เฉลี่ย:", average)

# เทียบ midterm กับ final ว่าใครทำข้อสอบ final ได้ดีขึ้น (ได้ array ของ True/False กลับมา)
improved = final > midterm
print("ทำได้ดีขึ้นไหม:", improved)
````

### 4.3 ฟังก์ชันสถิติพื้นฐานของ NumPy

NumPy มีฟังก์ชันคำนวณสถิติพื้นฐานในตัว เร็วกว่าเขียน loop เองมาก


````python
scores = np.array([78, 85, 62, 90, 55, 73, 88])

print("ผลรวม:", np.sum(scores))           # หรือ scores.sum()
print("ค่าเฉลี่ย:", np.mean(scores))       # หรือ scores.mean()
print("ค่าสูงสุด:", np.max(scores))        # หรือ scores.max()
print("ค่าต่ำสุด:", np.min(scores))        # หรือ scores.min()
print("ส่วนเบี่ยงเบนมาตรฐาน:", np.std(scores))   # standard deviation
print("ค่ามัธยฐาน:", np.median(scores))
print("เรียงลำดับ:", np.sort(scores))
````

### 4.4 Boolean Indexing — กรองข้อมูลด้วยเงื่อนไข (เชื่อมกับ if ที่เรียนมา)

จุดที่ทรงพลังมาก: ใส่เงื่อนไขเข้าไปใน `[]` ของ array เพื่อ "กรอง" เอาเฉพาะค่าที่ตรงเงื่อนไขออกมา — ทำงานคล้าย list comprehension ที่มี `if` แต่เร็วกว่ามากกับข้อมูลขนาดใหญ่


````python
scores = np.array([78, 85, 62, 90, 55, 73, 88])

# สร้าง boolean array จากเงื่อนไข (True/False ตามตำแหน่ง)
passing = scores >= 70
print("ผ่านไหม (boolean array):", passing)

# ใช้ boolean array นั้นกรองค่าออกมา (ได้เฉพาะค่าที่ True)
passing_scores = scores[scores >= 70]
print("คะแนนที่ผ่าน:", passing_scores)

# เทียบกับวิธีเดิมด้วย list comprehension
scores_list = [78, 85, 62, 90, 55, 73, 88]
passing_list = [s for s in scores_list if s >= 70]
print("เทียบกับ list comprehension:", passing_list)

````

````python
# นับจำนวนคนที่ผ่าน
print("จำนวนคนที่ผ่าน:", np.sum(scores >= 70))   # True นับเป็น 1, False นับเป็น 0
print("เปอร์เซ็นต์คนที่ผ่าน:", np.mean(scores >= 70) * 100, "%")
````

### 5.1 การคำนวณตามแถว/คอลัมน์ด้วย parameter `axis`

### 5.2 สรุป: numpy array vs list ปกติ

| | list ปกติ | numpy array |
|---|---|---|
| คำนวณกับค่าคงที่ (`+`, `*`) | ❌ ทำตรงๆไม่ได้ ต้องใช้ comprehension | ✅ ทำได้ตรงๆ (vectorized) |
| ความเร็วกับข้อมูลจำนวนมาก | ช้ากว่า | เร็วกว่ามาก |
| เก็บข้อมูลคนละชนิดในตัวเดียว | ✅ ทำได้ | ❌ ควรเป็นชนิดเดียวกันทั้งหมด |
| ฟังก์ชันสถิติสำเร็จรูป | ❌ ไม่มี ต้องเขียนเอง | ✅ มีให้พร้อม (`mean`, `std`, ...) |
| ใช้กับ pandas/scikit-learn | ใช้ได้บางส่วน | ✅ เป็นพื้นฐานหลักของทั้งวงการ Data Science |


## Matrix

### การคำนวณเกี่ยวกับ Matrix

````python
A = np.array([
    [1, 2],
    [3, 4]
])

B = np.array([
    [5, 6],
    [7, 8]
])

print("\nMatrix A:\n", A)
print("Matrix B:\n", B)

````

````python
# 1. การบวก/ลบ Matrix (Element-wise)
print("\nAddition (A + B):\n", A + B)
print("Subtraction (A - B):\n", A - B)
````

````python
# 2. การคูณ Matrix
#   a) Element-wise multiplication (คูณกันทีละตำแหน่ง)
print("\nElement-wise Multiplication (A * B):\n", A * B)
````

````python
#   b) Matrix multiplication (Dot product) - ใช้ @ หรือ np.dot()
#      จำนวนคอลัมน์ของ Matrix แรกต้องเท่ากับจำนวนแถวของ Matrix หลัง
C = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
D = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

print("\nMatrix C (2x3):\n", C)
print("Matrix D (3x2):\n", D)

matrix_product = C @ D
````

````python
# หรือ matrix_product = np.dot(C, D)
print("Matrix Multiplication (C @ D):\n", matrix_product)
````

````python
# 3. Transpose (สลับแถวเป็นคอลัมน์)
print("\nTranspose of A (A.T):\n", A.T)
````

### ข้อ 2: NumPy — 2D array และ axis
มีคะแนนสอบเก็บเป็น 2D array (4 คน x 3 วิชา):
```python
scores = np.array([
    [85, 70, 90],
    [60, 75, 65],
    [95, 88, 92],
    [55, 60, 58],
])
```

จงหา:
1. คะแนนเฉลี่ยของแต่ละวิชา (axis ไหน?)
2. คะแนนเฉลี่ยของแต่ละคน (axis ไหน?)
3. ใครได้คะแนนเฉลี่ยสูงสุด (Hint: ใช้ `np.mean(..., axis=...)` ร่วมกับ `np.argmax()` เพื่อหา index ของค่าสูงสุด)


````python
import numpy as np
scores = np.array([
    [85, 70, 90],
    [60, 75, 65],
    [95, 88, 92],
    [55, 60, 58],
])

# 1. คะแนนเฉลี่ยของแต่ละวิชา (axis=0)
subject_averages = np.mean(scores, axis=0)
print("คะแนนเฉลี่ยของแต่ละวิชา:", subject_averages)
````

````python
# 2. คะแนนเฉลี่ยของแต่ละคน (axis=1)
person_averages = np.mean(scores, axis=1)
print ("คะแนนเฉลี่ยของแต่ละคนคือ:", person_averages)
````

````python
# 3. ใครได้คะแนนเฉลี่ยสูงสุด
highest_avg_person_index = np.argmax(person_averages)
highest_avg_score = person_averages[highest_avg_person_index]

print(f"คนที่ได้คะแนนเฉลี่ยสูงสุดคือคนที่ {highest_avg_person_index + 1} ด้วยคะแนนเฉลี่ย {highest_avg_score:.2f}")
````

### ข้อ 3: ท้าทายเพิ่มเติม (Challenge) — รวม Function + OOP + NumPy
สร้าง class `Course` ที่มี:
1. `__init__(self, course_name)` — attribute `course_name` และ `scores` (list ว่างเริ่มต้น)
2. method `add_score(self, score)` — เพิ่มคะแนนเข้า `self.scores`
3. method `get_statistics(self)` — แปลง `self.scores` เป็น NumPy array แล้ว **return** dict ที่มี key: `mean`, `max`, `min`, `std` (ใช้ฟังก์ชัน NumPy ที่เรียนมา)
4. เขียนฟังก์ชัน `print_course_report(course)` (function ธรรมดา ไม่ใช่ method) ที่รับ object `Course` มา แล้ว print รายงานสรุปสถิติออกมาให้อ่านง่าย

ทดสอบสร้าง course เพิ่มคะแนน 5-6 ค่า แล้วเรียกดูรายงาน


````python

    def __init__(self, course_name):
        self.course_name = course_name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)
````

````python
import numpy as np

class Course:
    def __init__(self, course_name):
        self.course_name = course_name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def get_statistics(self):
        if not self.scores:
            return {"mean": 0.0, "max": 0.0, "min": 0.0, "std": 0.0}

        np_scores = np.array(self.scores)
        return {
            "mean": np.mean(np_scores),
            "max": np.max(np_scores),
            "min": np.min(np_scores),
            "std": np.std(np_scores)
        }
````

````python
def print_course_report(course):
    print(f"--- Course Report: {course.course_name} ---")
    if not course.scores:
        print("No scores recorded yet.")
        return

    stats = course.get_statistics()
    print(f"Total scores: {len(course.scores)}")
    print(f"Mean score: {stats['mean']:.2f}")
    print(f"Max score: {stats['max']:.2f}")
    print(f"Min score: {stats['min']:.2f}")
    print(f"Standard Deviation: {stats['std']:.2f}")
    print("------------------------------------")
````

````python
# สร้าง object ของ Course
my_course = Course("Data Science Fundamentals")

# เพิ่มคะแนน
my_course.add_score(85)
my_course.add_score(92)
my_course.add_score(78)
my_course.add_score(65)
my_course.add_score(90)
my_course.add_score(88)

# เรียกใช้ฟังก์ชัน print_course_report
print_course_report(my_course)
````
